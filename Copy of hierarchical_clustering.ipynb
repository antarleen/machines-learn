# Hierarchical Clustering

## Importing the libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,3:5].values
print(X)
```

## Using the dendrogram to find the optimal number of clusters

```
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```
from sklearn.cluster import AgglomerativeClustering
hc_1 = AgglomerativeClustering(n_clusters = 5,affinity = 'euclidean',linkage ='ward')
y_hc_1 = hc_1.fit_predict(X)
print(y_hc_1)
hc_2 = AgglomerativeClustering(n_clusters = 3,affinity = 'euclidean',linkage ='ward')
y_hc_2 = hc_2.fit_predict(X)
print(y_hc_2)
```

## Visualising the clusters

```
plt.scatter(X[y_hc_1 ==0,0],X[y_hc_1 == 0,1],s=100,c='red',label = 'Cluster 1')
plt.scatter(X[y_hc_1 ==1,0],X[y_hc_1 == 1,1],s=100,c='blue',label = 'Cluster 2')
plt.scatter(X[y_hc_1 ==2,0],X[y_hc_1 == 2,1],s=100,c='yellow',label = 'Cluster 3')
plt.scatter(X[y_hc_1 ==3,0],X[y_hc_1 == 3,1],s=100,c='green',label = 'Cluster 4')
plt.scatter(X[y_hc_1 ==4,0],X[y_hc_1 == 4,1],s=100,c='cyan',label = 'Cluster 5')
plt.title('Scatter for Customers')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.show()
```

```
plt.scatter(X[y_hc_2 ==0,0],X[y_hc_2 == 0,1],s=100,c='red',label = 'Cluster 1')
plt.scatter(X[y_hc_2 ==1,0],X[y_hc_2 == 1,1],s=100,c='blue',label = 'Cluster 2')
plt.scatter(X[y_hc_2 ==2,0],X[y_hc_2 == 2,1],s=100,c='yellow',label = 'Cluster 3')
plt.title('Scatter for Customers')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.show()
```
